```python
import csv

d=[
    ["name","email","mobile"],
    ["Abc","email@some.com","998899"],
]

with open ("data.csv","w",newline='') as file:
    writer = csv.writer(file)
    for i in d:
        writer.writerow(i)
```

```python
with open("data.csv","r") as file:
    reader = csv.reader(file)

    for row in reader:
        print(row)
```
