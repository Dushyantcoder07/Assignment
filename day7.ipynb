```python
import pandas as pd

df= pd.DataFrame({'A':[1,2,3,4,5],'B':[10,20,30,40,50]})

print(df)
print("\n Addition\n",df['A'][0]+5)
print("\n Subtraction\n",df-5)
```

```python
import pandas as pd

df1= pd.DataFrame({'A':[1,2,3,4],'B':[10,20,30,40]})
df2= pd.DataFrame({'A':[11,22,33],'B':[12,25,36]}, index=[1,2,3])


print(df1)
print(df2)
print("\n Addition\n",df1+df2)
print("\nSubstraction", df1-df2)
print("\nMultiplication\n",df1*df2)
```

```python
import pandas as pd

one = pd.DataFrame({
    "Name": ["b1","b2","b3"],
    "subject": ["1","2","3"],
    "Marks":[10,20,30]
}, index=[0,1,2])

two = pd.DataFrame({
    "Name": ["A1","A2","A3"],
    "subject": ["4","5","6"],
    "Marks":[23,40,50]
}, index=[0,1,2])
res= pd.concat([one,two])
print(res)
```
