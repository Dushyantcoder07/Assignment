```python
import csv

# File name
filename = "address_book.csv"

# Column headers
fields = ["Name", "Address", "Mobile", "Email"]

# Empty list to store user data
rows = []

# Taking input for 3 users
for i in range(3):
    print(f"\nEnter details for Person {i+1}")
    
    name = input("Enter Name: ")
    address = input("Enter Address: ")
    mobile = input("Enter Mobile Number: ")
    email = input("Enter Email: ")
    
    rows.append([name, address, mobile, email])

# Writing data into CSV file
with open(filename, mode="w", newline="") as file:
    writer = csv.writer(file)
    
    # Write headers
    writer.writerow(fields)
    
    # Write user data
    writer.writerows(rows)

print(f"\nCSV file '{filename}' created successfully!")
```
